```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

irisDataset = pd.read_csv('./datasets/iris.csv', index_col="Id")
irisDataset.head(10)
```

```python
plt.figure(figsize=(20, 6))
sns.histplot(irisDataset["Petal Length (cm)"])
```

```python
plt.figure(figsize=(20, 6))
sns.kdeplot(data=irisDataset['Petal Length (cm)'], shade=True)
```

```python
plt.figure(figsize=(20, 6))
sns.jointplot(x=irisDataset["Petal Length (cm)"], y=irisDataset["Sepal Width (cm)"], kind='kde', shade=True, color='g')
```

```python
plt.figure(figsize=(20, 6))
sns.histplot(data=irisDataset, x='Petal Length (cm)', hue='Species')
plt.title('Histogram of Petal Lengths, by Species')
```

```python
plt.figure(figsize=(20, 6))
sns.kdeplot(data=irisDataset, x='Petal Length (cm)', hue='Species', shade=True)
plt.title('Histogram of Petal Lengths, by Species')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

flightFilePath = './datasets/flight_delays.csv'
flightDataset = pd.read_csv(flightFilePath, index_col='Month')

flightDataset.head(10)
```

```python
plt.figure(figsize=(20, 6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")
plt.ylabel("Arrival Delay(per min)")
sns.barplot(x=flightDataset.index, y=flightDataset['NK'])
```

```python
plt.figure(figsize=(20, 6))
plt.title("Arrival Delay for Each Airline, by Month")
#annotation(actual delay time on each cell) is shown
sns.heatmap(data=flightDataset, annot=True)
plt.xlabel("Airlines")
```

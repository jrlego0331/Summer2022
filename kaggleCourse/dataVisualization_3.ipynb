```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

insuranceDataset = pd.read_csv("./datasets/insurance.csv")
insuranceDataset.head(10)
```

```python
plt.figure(figsize=(20, 6))
sns.scatterplot(x=insuranceDataset['bmi'], y=insuranceDataset['charges'])
```

```python
plt.figure(figsize=(20, 6))
sns.regplot(x=insuranceDataset['bmi'], y=insuranceDataset['charges'])
```

```python
plt.figure(figsize=(20, 6))
sns.scatterplot(x=insuranceDataset['bmi'], y=insuranceDataset['charges'], hue=insuranceDataset['smoker'])
```

```python
plt.figure(figsize=(20, 6))
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insuranceDataset)
```

```python
plt.figure(figsize=(20, 6))
sns.swarmplot(x='smoker', y='charges', data=insuranceDataset)
```

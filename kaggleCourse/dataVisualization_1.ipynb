```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

spotifyFilePath = './datasets/spotify.csv'
spotifyDataset = pd.read_csv(spotifyFilePath, index_col='Date', parse_dates=True)
spotifyDataset.head(10)
```

```python
plt.figure(figsize=(20,6))
sns.lineplot(data=spotifyDataset)
plt.title('Daily Global Streams of Popular Songs 2017-2018')
```
